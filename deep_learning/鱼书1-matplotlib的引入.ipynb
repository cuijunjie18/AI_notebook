<strong>matplotlib的引入及简单应用</strong>

```python
#导入第三方库
import numpy as np
import matplotlib.pyplot as plt
```

```python
#生成数据

x = np.arange(0,6,0.1)#生成0到6(不包含6),以0.1为间隔的数组
y = np.sin(x)#计算正弦值,本质为矩阵运算

#绘图
plt.plot(x,y)#第一个参数为自变量,第二个参数为因变量
plt.show()
```

```python
#plt的其他功能

x = np.arange(0,6,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

#在一幅图上绘制两个曲线
plt.plot(x,y1,label = 'sin')
plt.plot(x,y2,label = 'cos',linestyle = '--')

#添加上面图例的label
plt.legend()

#设置x,y坐标及标题
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin&cos')

plt.show()
```
